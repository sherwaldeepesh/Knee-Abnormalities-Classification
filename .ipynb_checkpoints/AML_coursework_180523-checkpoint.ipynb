```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import timm
```

```python
class CustomDataset(Dataset):
    def __init__(self, npy_array, labels, transform=None):
        self.npy_array = npy_array
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.npy_array)

    def __getitem__(self, idx):
        image = self.npy_array[idx]
        label = self.labels[idx]
        
        if self.transform:
            image = self.transform(image)
        
        return image, label
```

```python
data_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.432], std=[0.501])
])
```

```python
dataset = CustomDataset(train_sagittal_npys, npy_label, transform=data_transform)
```

```python
batch_size = 32
shuffle = True

data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)
```

```python
# model = timm.create_model('vit_base_patch16_224', pretrained=True)
model = timm.create_model('vit_small_patch16_224.augreg_in21k_ft_in1k', pretrained=True, num_classes = 0)
model.head = nn.Sequential(
    nn.BatchNorm1d(384),
    nn.Linear(in_features=384, out_features=64, bias=False),
    nn.ReLU(),
    # nn.BatchNorm1d(64),
    # nn.Dropout(0.4),
    nn.Linear(in_features=64, out_features=2, bias=False)
)
```

```python
# num_classes = 2  # Assuming you have 2 classes (e.g., abnormal and normal)
# model.head = nn.Linear(model.head.in_features, num_classes)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)
```

```python
# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# model.to(device)

# num_epochs = 10

# for epoch in range(num_epochs):
#     for images, labels in data_loader:
#         images = images.to(device)
#         labels = labels.to(device)
        
#         optimizer.zero_grad()
        
#         outputs = model(images)
#         loss = criterion(outputs, labels)
        
#         loss.backward()
#         optimizer.step()
        
#     scheduler.step()
#     print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
# from sklearn.metrics import confusion_matrix, classification_report

# device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
# model.to(device)
# num_epochs = 10

# for epoch in range(num_epochs):
#     # Initialize variables for tracking predictions and labels
#     all_predictions = []
#     all_labels = []
    
#     for images, labels in data_loader:
#         images = images.to(device)
#         labels = labels.to(device)
        
#         optimizer.zero_grad()
        
#         outputs = model(images)
#         loss = criterion(outputs, labels)
        
#         loss.backward()
#         optimizer.step()
        
#         # Collect predictions and labels
#         _, predictions = torch.max(outputs, 1)
#         all_predictions.extend(predictions.cpu().tolist())
#         all_labels.extend(labels.cpu().tolist())
        
#     scheduler.step()
#     print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
    
#     # Compute and print confusion matrix
#     cm = confusion_matrix(all_labels, all_predictions)
#     print("Confusion Matrix:")
#     print(cm)
    
#     # Compute and print classification report
#     report = classification_report(all_labels, all_predictions)
#     print("Classification Report:")
#     print(report)

from sklearn.metrics import confusion_matrix, classification_report

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Rest of the code...

num_epochs = 10

for epoch in range(num_epochs):
    # Initialize variables for tracking predictions and labels
    all_predictions = []
    all_labels = []
    
    for images, labels in data_loader:
        images = images.to(device)
        labels = labels.to(device)
        
        optimizer.zero_grad()
        
        outputs = model(images)
        loss = criterion(outputs, labels)
        
        loss.backward()
        optimizer.step()
        
        # Collect predictions and labels
        _, predictions = torch.max(outputs, 1)
        all_predictions.extend(predictions.cpu().tolist())
        all_labels.extend(labels.cpu().tolist())
        
    scheduler.step()
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")
    
    # Compute and print confusion matrix
    cm = confusion_matrix(all_labels, all_predictions)
    print("Confusion Matrix:")
    print(cm)
    
    # Compute and print classification report
    report = classification_report(all_labels, all_predictions)
    print("Classification Report:")
    print(report)
    
    # Plot confusion matrix with color
    class_names = ['Class 0', 'Class 1']  # Replace with your class names
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='d', xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
import numpy as np
import os
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
import timm
import matplotlib.pyplot as plt


# Path to the folder containing the 3D knee images
folder_path = r'D:\AML Course-work\data\MRNet-v1.0\train\sagittal'
#folder_path ='/content/drive/MyDrive/Surrey/CourseWorkAML/data/valid/axial'

# Initialize an empty list to store the resized 2D NPY arrays
train_sagittal_npys = []

# Iterate over each file in the folder
for file_name in os.listdir(folder_path):
    if file_name.endswith('.npy'):  # Assuming the files are in .npy format
        file_path = os.path.join(folder_path, file_name)
        
        # Load the 3D knee image as a NumPy array
        knee_image = np.load(file_path)
        
        # Select the desired slice from the 3D image (similar to the previous example)
        slice_index = 15
        axis = 0
        slice_2d = knee_image.take(indices=slice_index, axis=axis)
        
        # Resize the 2D slice to 224x224 pixels
        resized_slice = Image.fromarray(slice_2d).resize((224, 224), Image.BILINEAR)
        
        # Convert the resized slice to RGB format (3 channels)
        resized_slice = resized_slice.convert('RGB')
        
        # Convert the resized slice to a NumPy array
        resized_slice = np.array(resized_slice)
        
        # Normalize the pixel values
        resized_slice = resized_slice / 255.0
        
        # Append the resized 2D slice to the list
        train_sagittal_npys.append(resized_slice)

# Convert the list of NPY arrays to a NumPy array
train_sagittal_npys = np.stack(train_sagittal_npys)

# Print the shape of each resized 2D NPY array
for npy_array in train_sagittal_npys:
    print(npy_array.shape)
```

```python
def plot_random(train_sagittal_npys, num=5):
    healthy_imgs = train_sagittal_npys[np.random.choice(train_sagittal_npys.shape[0], num, replace=False)]
        
    plt.figure(figsize=(16,9))
    for i in range(num):
        plt.subplot(1, num, i+1)
        plt.title('Knee Abnormal')
        plt.imshow(healthy_imgs[i])
        
plot_random(train_sagittal_npys, num=5)  
```

```python
from torch.utils.data import DataLoader
# Create an instance of the custom dataset
#dataset = CustomDataset(train_sagittal_npys, npy_lable, transform=data_transform)


# from torchvision import transforms

# class NumpyToPIL(object):
#     def __call__(self, numpy_array):
#         # Convert NumPy array to PIL image
#         image = Image.fromarray(numpy_array)
#         return image

# # Define data augmentations
# data_transform = transforms.Compose([
#     NumpyToPIL(),                        # Convert NumPy array to PIL image
#     transforms.RandomRotation(10),       # Randomly rotate the image by 10 degrees
#     transforms.RandomHorizontalFlip(),   # Randomly flip the image horizontally
#     transforms.ToTensor(),                # Convert the image to tensor
#     transforms.Normalize(mean=[0.5], std=[0.5])   # Normalize the image
# ])

# # Create an instance of the custom dataset with data augmentations
# dataset = CustomDataset(train_sagittal_npys, npy_lable, transform=data_transform)

from torchvision import transforms

class NumpyToPIL(object):
    def __call__(self, numpy_array):
        # Check the shape of the array
        if len(numpy_array.shape) == 2:  # Grayscale image
            # Convert grayscale image to PIL image
            image = Image.fromarray(numpy_array, mode='L')
        elif len(numpy_array.shape) == 3:  # RGB image
            # Convert RGB image to PIL image
            image = Image.fromarray(numpy_array, mode='RGB')
        else:
            raise ValueError("Unsupported array shape")

        return image

# Define data augmentations
data_transform = transforms.Compose([
    NumpyToPIL(),                        # Convert NumPy array to PIL image
    transforms.RandomRotation(10),       # Randomly rotate the image by 10 degrees
    transforms.RandomHorizontalFlip(),   # Randomly flip the image horizontally
    transforms.ToTensor(),                # Convert the image to tensor
    transforms.Normalize(mean=[0.5], std=[0.5])   # Normalize the image
])

# Create an instance of the custom dataset with data augmentations
dataset = CustomDataset(train_sagittal_npys, npy_lable, transform=data_transform)

# Create a data loader for batching and shuffling the data
batch_size = 32
shuffle = True
data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)



```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)
print(device)

num_epochs = 2
learning_rate = 0.001
weight_decay = 1e-5
batch_size = 16  # Define the batch size
shuffle = True  # Specify whether to shuffle the data

model = model.double()

optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)

scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.1)

# Create a data loader for batching and shuffling the data



for epoch in range(num_epochs):
    for images, labels in data_loader:
        images = images.to(device).double()
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        outputs = outputs.view(-1, num_classes)  # Reshape to (batch_size, num_classes)

        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

    scheduler.step()
    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# for epoch in range(num_epochs):
#     for images, labels in data_loader:
#         images = images.to(device).double()
#         labels = labels.to(device)

#         optimizer.zero_grad()

#         outputs = model(images)
#         outputs = outputs.view(-1)  # Reshape to (batch_size,)

#         loss = criterion(outputs, labels.float())

#         loss.backward()
#         optimizer.step()

#     scheduler.step()
#     print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}")

# Evaluation code
model.eval()
predictions = []
true_labels = []

with torch.no_grad():
    for images, labels in data_loader:
        images = images.to(device).double()
        labels = labels.to(device)

        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)

        predictions.extend(predicted.cpu().numpy().tolist())  # Convert to Python list
        true_labels.extend(labels.cpu().numpy().tolist())  # Convert to Python list

# Convert predictions and true labels to NumPy arrays
predictions = np.array(predictions)
true_labels = np.array(true_labels)

# Print classification report
print("Classification Report:")
print(classification_report(true_labels, predictions))

# Calculate confusion matrix
cm = confusion_matrix(true_labels, predictions)

# Plot confusion matrix with colors
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision.transforms import ToTensor
from sklearn.model_selection import train_test_split
!pip install einops
```

```python
class MRNetDataset(Dataset):
    def __init__(self, file_paths, labels, transform=None):
        self.file_paths = file_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.file_paths)

    def __getitem__(self, index):
        file_path = self.file_paths[index]
        label = self.labels[index]

        try:
            scan_data = np.load(file_path)
            scan_data = scan_data.astype(np.float32)

            # Convert 3D scan to 2D axial slices
            axial_slices = scan_data.shape[2]
            scan_data_2d = []
            for slice_idx in range(axial_slices):
                slice_data = scan_data[:, :, slice_idx]
                scan_data_2d.append(slice_data)
            scan_data_2d = np.stack(scan_data_2d, axis=0)

            if self.transform:
                scan_data_2d = self.transform(scan_data_2d)

            return scan_data_2d, label

        except Exception as e:
            print(f"Error loading file: {file_path}")
            print(f"Error message: {str(e)}")
            return None, None
```

```python
# Read the CSV file and format the file names
folder_path = '/content/drive/MyDrive/Surrey/CourseWorkAML/data/train/axial'
df = pd.read_csv(os.path.join(folder_path, 'train-abnormal.csv'), header=None, names=['file_number', 'label'])
df['file_number'] = df['file_number'].apply(lambda x: str(x).zfill(4))
df['file_path'] = folder_path + '/' + df['file_number'] + '.npy'

# Split the data into training and validation sets
train_files, valid_files, train_labels, valid_labels = train_test_split(
    df['file_path'], df['label'], test_size=0.2, random_state=42)
```

```python
from torchvision.transforms import ToTensor
from torch.utils.data import DataLoader

# Define data transformations
transform = ToTensor()  # You can add more transformations as needed

# Create data loaders
train_dataset = MRNetDataset(train_files, train_labels, transform=transform)
valid_dataset = MRNetDataset(valid_files, valid_labels, transform=transform)

batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from einops.layers.torch import Rearrange

class VisionTransformer(nn.Module):
    def __init__(self, patch_size, num_classes):
        super(VisionTransformer, self).__init__()

        self.patch_size = patch_size
        self.num_classes = num_classes

        self.patch_embedding = nn.Conv2d(1, 256, kernel_size=patch_size, stride=patch_size)
        self.rearrange = Rearrange('b c h w -> b (h w) c')
        self.transformer = nn.Transformer(d_model=256, nhead=8, num_encoder_layers=6)
        self.fc = nn.Linear(256, num_classes)

    def forward(self, x):
        x = self.patch_embedding(x)
        x = self.rearrange(x)
        x = x.permute(1, 0, 2)  # (seq_len, batch_size, emb_dim)
        x = self.transformer(x)
        x = x.mean(dim=0)  # Average pooling over the sequence length
        x = self.fc(x)
        return x
```

```python
patch_size = 32  # Set the patch size according to your preference
num_classes = 2  # Set the number of classes

model = VisionTransformer(patch_size, num_classes)
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

model.to(device)

num_epochs = 10  # Set the number of training epochs

for epoch in range(num_epochs):
    model.train()
    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = F.cross_entropy(outputs, labels)

        loss.backward()
        optimizer.step()

    model.eval()
    total_correct = 0
    total_samples = 0
    with torch.no_grad():
        for images, labels in valid_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            _, predicted = torch.max(outputs, 1)

            total_samples += labels.size(0)
            total_correct += (predicted == labels).sum().item()

    accuracy = total_correct / total_samples
    print(f'Epoch [{epoch+1}/{num_epochs}], Validation Accuracy: {accuracy:.4f}')
```

```python
import os
import pandas as pd
import numpy as np
import torch
import torchvision.transforms as transforms
import torch.utils.data as data

class MRDataset(data.Dataset):
    def __init__(self, root_dir, task, plane, train=True, transform=None, weights=None):
        super().__init__()
        self.task = task
        self.plane = plane
        self.root_dir = root_dir
        self.train = train
        
        if self.train:
            self.folder_path = os.path.join(self.root_dir, 'train', self.plane)
            self.records = pd.read_csv(os.path.join(self.root_dir, f'train-{task}.csv'), header=None, names=['id', 'label'])
        else:
            self.transform = None
            self.folder_path = os.path.join(self.root_dir, 'valid', self.plane)
            self.records = pd.read_csv(os.path.join(self.root_dir, f'valid-{task}.csv'), header=None, names=['id', 'label'])

        self.records['id'] = self.records['id'].apply(lambda i: str(i).zfill(4))
        self.paths = [os.path.join(self.folder_path, f'{filename}.npy') for filename in self.records['id'].tolist()]
        self.labels = self.records['label'].tolist()

        self.transform = transform
        if weights is None:
            pos = np.sum(self.labels)
            neg = len(self.labels) - pos
            self.weights = torch.FloatTensor([1, neg / pos])
        else:
            self.weights = torch.FloatTensor(weights)

    def __len__(self):
        return len(self.paths)

    def __getitem__(self, index):
        array = np.load(self.paths[index])
        label = self.labels[index]
        
        if label == 1:
            label = torch.FloatTensor([0, 1])
        else:
            label = torch.FloatTensor([1, 0])

        if self.transform:
            array = self.transform(array)
        else:
            array = torch.from_numpy(array).unsqueeze(0)  # Add channel dimension

        return array, label, self.weights

# Set the root directory where the dataset is located
root_dir = '/content/drive/MyDrive/Surrey/CourseWorkAML/data/train'

# Set the task and plane
task = 'abnormal'  # Choose from ['abnormal', 'acl', 'meniscus']
plane = 'sagittal'  # Choose from ['sagittal', 'coronal', 'axial']

# Define data transformations
transform = transforms.Compose([
    # Add your desired transformations here
    transforms.ToTensor()
])

# Create the dataset
train_dataset = MRDataset(root_dir, task, plane, train=True, transform=transform)
valid_dataset = MRDataset(root_dir, task, plane, train=False, transform=transform)

# Create data loaders
batch_size = 32
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader = torch.utils.data.DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)

# Now you can use the train_loader and valid_loader for training and validation
```
